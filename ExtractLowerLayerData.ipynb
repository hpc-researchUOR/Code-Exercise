```python
from pathlib import Path
import json
import pandas as pd

root = Path.cwd()


def get_attack_class(path):
    path_str = str(path).lower()
    if "bruteforce" in path_str:
        return "BruteForce"
    if "ddos" in path_str:
        return "DDOS"
    if "dos" in path_str and "ddos" not in path_str:
        return "DoS"
    if "benign" in path_str:
        return "Benign"
    return "Unknown"


def get_text_attack_class(text_path):
    return get_attack_class(text_path)
```

```python
from pathlib import Path
import json
import pandas as pd

root = Path.cwd()
output_root = root / "csv_output"
output_root.mkdir(parents=True, exist_ok=True)

def save_dataframe(df, path):
    path.parent.mkdir(parents=True, exist_ok=True)
    df.to_csv(path, index=False)
    print(f"Saved {len(df)} rows to {path}")

def extract_lower_layer_texts():
    text_paths = sorted(root.rglob("**/Lower_Layer/*.txt"))
    if not text_paths:
        print("No Lower_Layer text files found.")
        return

    all_rows = []
    for text_path in text_paths:
        rows = []
        attack_class = get_attack_class(text_path)
        with text_path.open("r", encoding="utf-8", errors="ignore") as f:
            for line in f:
                line = line.strip()
                if not line:
                    continue
                row = json.loads(line)
                row["attack_class"] = attack_class
                rows.append(row)

        if not rows:
            continue

        df = pd.DataFrame(rows)
        csv_path = output_root / "lower_layer" / text_path.relative_to(root).with_suffix(".csv")
        save_dataframe(df, csv_path)
        all_rows.extend(rows)

    if all_rows:
        save_dataframe(pd.DataFrame(all_rows), output_root / "lower_layer_all.csv")

if __name__ == "__main__":
    extract_lower_layer_texts()
```
