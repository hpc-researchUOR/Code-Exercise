```python
# Import one ddos data set
df3 = pd.read_csv("new_csv_output/ddos_icmp_hping3.csv")
df3.head(5)
```

```python
# Import one dos data set
df4 = pd.read_csv("new_csv_output/icmp_hping3.csv")
df4.head(5)
```

```python
dfs = {
    "df1": df1,
    "df2": df2,
    "df3": df3,
    "df4": df4
}

reference = list(df1.columns)

for name, df in dfs.items():
    if list(df.columns) == reference:
        print(f"{name}: Columns match")
    else:
        print(f"{name}: Columns do NOT match")
```

```python
import pandas as pd

# Add Attack Type column
df1["Attack Type"] = "Benign"
df2["Attack Type"] = "DDoS"
df3["Attack Type"] = "DOS"
df4["Attack Type"] = "Bruteforce"

# Merge all DataFrames
merged_df = pd.concat(
    [df1, df2, df3, df4],
    ignore_index=True
)

# Save to CSV
merged_df.to_csv("merged_dataset.csv", index=False)

# Check the result
print("Merged dataset shape:", merged_df.shape)
print("\nAttack Type counts:")
print(merged_df["Attack Type"].value_counts())
```

```python
# TCP packets with missing ports
tcp_missing = df1[(df1["protocol"] == 6) &
                 (df1["src_port"].isna() | df1["dst_port"].isna())]

print("TCP packets missing ports:", len(tcp_missing))

# UDP packets missing UDP length
udp_missing = df1[(df1["protocol"] == 17) &
                 (df1["udp_length"].isna())]

print("UDP packets missing udp_length:", len(udp_missing))

# ICMP packets missing type/code
icmp_missing = df1[(df1["protocol"] == 1) &
                  (df1["icmp_type"].isna() | df1["icmp_code"].isna())]

print("ICMP packets missing fields:", len(icmp_missing))
```

#### Fill structural missing values

Instead of deleting columns or rows, replace them with a sentinel value.

```python
numeric_cols = [
    "src_port",
    "dst_port",
    "seq",
    "ack",
    "window",
    "udp_length",
    "icmp_type",
    "icmp_code"
]

df1[numeric_cols] = df1[numeric_cols].fillna(-1)
```

```python
categorical_cols = [
    "src_mac",
    "dst_mac",
    "eth_type"
]

df1[categorical_cols] = df1[categorical_cols].fillna("Unknown")
```

```python
df1["is_tcp"] = (df1["protocol"] == 6).astype(int)
df1["is_udp"] = (df1["protocol"] == 17).astype(int)
df1["is_icmp"] = (df1["protocol"] == 1).astype(int)
```
